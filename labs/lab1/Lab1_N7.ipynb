# Задание
Сгенерировать __N__ облаков (классов) с __M__ точками (объектами) в каждом. Построить классификатор используя *scikit-learn* который достигнет точности __R__ лежащей в интервале __(R1, R2)__.

----
Generate <b>N</b> clouds (classes) with <b>M</b> points (samples) in each cloud. Build classificator (SKLearn) which yields an accuracy <b>R</b> within the interval <b>(R1, R2)</b>. <br>

Группа 6132

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
```

```python
N = 7
M = 775
R1 = 0.705
R2 = 0.712
```

```python
accuracy = 0   # Точность 
step = 0       # Сдвиг 

while accuracy < R1:             
    step += 0.01

    np.random.seed(0)          # Начальное условие для генерации

    # Данные генерируются нормальным распределением и сдвигаются друг от друга
    cloud1 = np.random.randn(M, 2)
    cloud2 = np.random.randn(M, 2) + step
    cloud3 = np.random.randn(M, 2) + 2*step
    cloud4 = np.random.randn(M, 2) + 3*step
    cloud5 = np.random.randn(M, 2) + 4*step
    cloud6 = np.random.randn(M, 2) + 5*step
    cloud7 = np.random.randn(M, 2) + 6*step

    # Вертикальное объединение массивов
    X = np.vstack((cloud1, cloud2, cloud3, cloud4, cloud5, cloud6, cloud7))

    # Классификация объектов массива X
    y = [0] * M + [1] * M + [2] * M + [3] * M + [4] * M + [5] * M + [6] * M 

    data, labels = shuffle(X, y)    # Случайная перестановка элементов массивов

    # Половина данных тренировочные, а другая для тестирования
    test_size = int(len(data) * 0.5)
    Xtrain = data[:test_size]
    Ytrain = labels[:test_size]
    Xtest = data[test_size:]
    Ytest = labels[test_size:]

    clf = LogisticRegression()
    clf.fit(Xtrain, Ytrain)   # Обучение

    # Предсказанная классификация тестовых данных
    predictions = clf.predict(Xtest)    
    # Вычисление точности
    accuracy = accuracy_score(Ytest, predictions)

print(f"Сдвиг {step}")
print(f"Точность = {accuracy}")
```
